```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
df = pd.read_csv('epa-sea-level.csv')
```

```python
b1, b0, r_2, p_value, std_err = linregress(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])

fig, ax = plt.subplots(figsize=(15, 6))
df.plot(kind='scatter', y='CSIRO Adjusted Sea Level', x='Year', ax=ax)

future_year = 2050
future_sea_level = b1*future_year+b0

plt.scatter(future_year, future_sea_level, color='red')

x_fit = np.linspace(min(df['Year']), future_year + 1, 100)

y_fit = b1 * x_fit + b0

plt.plot(x_fit, y_fit, label='Line of Best Fit', color='purple')

start_date = 2000

# df_new = df[df['Year'] >= start_date]

b1_1, b0_1, r_2_1, p_value_1, std_err_1 = linregress(x=df[df['Year'] >= start_date]['Year'], y=df[df['Year'] >= start_date]['CSIRO Adjusted Sea Level'])

# fig, ax = plt.subplots(figsize=(15, 6))
df[df['Year'] >= start_date].plot(kind='scatter', y='CSIRO Adjusted Sea Level', x='Year', ax=ax)

future_year_1 = 2050
future_sea_level_1 = b1_1*future_year_1+b0_1

plt.scatter(future_year_1, future_sea_level_1, color='blue')

x_fit_1 = np.linspace(start_date, future_year_1 + 1, 100)

y_fit_1 = b1_1 * x_fit + b0_1

plt.plot(x_fit_1, y_fit_1, label='Line of Best Fit', color='red')

# plt.show()

plt.show()
```
